# Basic plots

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#matplotlib inline is used for visualiztion of graphs in the jupyter notebook itself otherwise we can use plt.show() to visualize graphs in separate window
```

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
#generate data
```

```python
fig = plt.figure() #create canvas for plotting graph
ax = fig.add_axes([0.1,0.1,1,1]) #add axes: need four argumnets[distance of x-axes, distance of y-axes, x-limit, y-limit]
ax.plot(x,y, c='red', alpha=0.8, marker='.', lw=2.5, label='Sine') 
ax.plot(x,z, c='blue', alpha=0.8, marker='x', lw=1.5, label='Cosine')
#c for colour, alpha for transparency, marker to mark different points on graph, lw for width of graph, label for labelling the name

ax.set_xlim(0,10) #to set xlim for graph
ax.set_ylim(-1,1)
ax.legend() #to display legends

ax.set_xlabel('Time') #to display labels
ax.set_ylabel('Amplitude')
ax.set_title('First matplotlib graph') #to display the title
```
